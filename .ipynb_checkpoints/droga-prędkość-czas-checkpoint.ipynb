# Droga, prędkość, czas

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
v = float(input("Podaj prędkość w km/h:"))
t = float(input("Podaj czas w godzinach:"))
```

```python
x = np.linspace (0, t, 20) 
s = []
for i in x:
    s.append(i*v)
plt.plot(x,s)
plt.title(r'Zależność drogi od czasu')
plt.xlabel('t [h]')
plt.ylabel('s [km]')
plt.grid()
plt.show()
```

# Prędkość światła

```python
v = 300000
t = 15
x = np.linspace (0, t, 20) 
s = []
for i in x:
    s.append(i*v/3600)
plt.plot(x,s)
plt.title(r'Zależność drogi od czasu dla prędkości światła')
plt.xlabel('t [s]')
plt.ylabel('s [km]')
plt.grid()
plt.show()
```

# Droga, prędkość, czas

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
v = float(input("Podaj prędkość w km/h:"))
t = float(input("Podaj czas w godzinach:"))
```

```python
x = np.linspace (0, t, 20) 
s = []
for i in x:
    s.append(i*v)
plt.plot(x,s,label='v')
plt.title('Zależność drogi od czasu')
plt.xlabel('t [h]')
plt.ylabel('s [km]')
plt.grid()
plt.legend()
plt.show()
```

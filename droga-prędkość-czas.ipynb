# Droga, prędkość, czas

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
v = float(input("Podaj prędkość w km/h:"))
t = float(input("Podaj czas w godzinach:"))
```

```python
x = np.linspace (0, t, 20) 
s = []
for i in x:
    s.append(i*v)
plt.plot(x,s)
plt.title(r'Zależność drogi od czasu')
plt.xlabel('t [h]')
plt.ylabel('s [km]')
plt.grid()
plt.show()
```

# Gepard

```python
v = 120
t = 1
x = np.linspace (0, t, 20) 
s = []
for i in x:
    s.append(i*v)
plt.plot(x,s)
plt.title(r'Zależność drogi od czasu dla Geparda')
plt.xlabel('t [h]')
plt.ylabel('s [km]')
plt.yticks([0, 20, 30, 40, 50, 60, 70, 80, 90, 100, 110, 120,])
plt.grid()
plt.show()
```
